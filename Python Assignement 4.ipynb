# 1b: Import the marriage census data into a DataFrame using replace method and prints the first ten rows.

```python
#the below code imports the  marriage census csv file into a dataframe
import pandas as pd
df = pd.DataFrame.from_csv(r'C:\Users\rithu\Downloads\census_marriage.csv', parse_dates=False)
df.head(10)
```

# 1c:Replaced the marriage status codes (marst column) with their actual status (e.g., Married, Single, etc.) using the replace method

```python
#the first list is the existing one and the second is the one it has to be replaced with
df.marst.replace([0,1,2,3,4,5,6],['Unknown','Married','Married/Spouse Absent','Separated','Divorced','Widowed','Single'],inplace=True)
df.head(10)
```

# 1d: Replaced the gender values with their gender equivalents (1 for Male, 2 for Female) using the replace method

```python
df.gender.replace([1,2],['Male','Female'],inplace=True)
df.head(10)
```

# 1e: Report the total number of divorced individuals in 1980 using DataFrame methods.

```python
#first reset the index and then set the two columns required as the index and got the required number using loc and used sum to get the total count
df=df.reset_index()
df_new= df.set_index(keys=['year','marst'])
```

```python
y=df_new.loc[(1980,'Divorced')]
```
