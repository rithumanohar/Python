```python
#importing pandas, regular expression and nan from numpy packages
import pandas as pd
import re
import numpy as np
from numpy import nan as NA
pd.set_option('display.max_colwidth', -1)
#data input as csv using pandas into a table and displayed to show how it looks
url=r'C:\Users\rithu\Desktop\Semester 2\Python\amazon_co-ecommerce_sample.csv'
table=pd.read_csv(url,index_col=None,header=None,encoding='utf-8-sig')
table.head(3)
```

```python
#using the below statement to load data so that the first row can be skipped directly through this statement
table=pd.read_csv(url,index_col=None,header=None,skiprows=1,encoding='utf-8-sig')
```

```python
#renaming the columns of importance
table.columns=['Unique ID','Product Name','Brand','Price(£)','Number Available In Stock','Number of Reviews','number_of_answered_questions','Average Review Rating','amazon_category_and_sub_category','customers_who_bought_this_item_also_bought','description','product_information','product_description','items_customers_buy_after_viewing_this_item','customer_questions_and_answers','customer_reviews','sellers']
```

```python
#assigning different numbers as 'Unique ID' as the given 'Unique' ID is ambiguous and long
table['Unique ID']=range(10000)
#using zfill to pad more zeroes preceding the number
table['Unique ID']=table['Unique ID'].apply(lambda x:str(x).zfill(4))
#using Unique ID as the index of the table
table.index=table['Unique ID']
```

```python
#to show the progress
table.head(3)
```

```python
#dropping all the unecessary columns from the table
table=table.drop(['Unique ID','number_of_answered_questions','amazon_category_and_sub_category','customers_who_bought_this_item_also_bought','description','product_description','customer_questions_and_answers','customer_reviews','items_customers_buy_after_viewing_this_item'],axis=1)
```

```python
#removing all the commas from the table using strip
table = table.applymap(lambda x: str(x).strip('\,'))
```

```python
#removing the '£' symbol from the 'Price(£)' column
table['Price(£)']=table['Price(£)'].map(lambda x: str(x).replace('£',''))
```

```python
#to show the progress
table.head(3)
```

```python
#removing unwanted text from the 'Average Review Rating' column
table['Average Review Rating']=table['Average Review Rating'].map(lambda x:str(x).strip(' out of 5 stars'))
```

```python
#removing the attached string from the number in the 'Number Available In Stock' column
table['Number Available In Stock']=table['Number Available In Stock'].map(lambda x:str(x).replace('new','').replace('used','').replace('collectible','').replace('refurbished','').replace('nan','0'))
```

```python
#extracting the 'Age Group' information from the 'production_information' using findall regular expression 
table['product_information']=table['product_information'].apply(lambda x:str(x))
table['Age Group']=table['product_information'].apply(lambda x:(re.findall('age:[(\d{1,2})]',x))[0:1])
table['Age Group']=table['Age Group'].map(lambda x:str(x).replace('age:',''))
```

```python
#to show the progress
table.head(3)
```

```python
#removing all the square brackets ([]),star(*) and apostophe(') from all the table using strip in lambda function
table = table.applymap(lambda x: str(x).strip('[]\*\''))
```

```python
#cleaning the sellers column to prepare it for extraction
table['sellers']=table['sellers'].apply(lambda x:str(x))
table['sellers']=table['sellers'].map(lambda x:x.replace('{"seller"=>','').replace('}','').replace('{','').replace('\"',''))
table['sellers']=table['sellers'].map(lambda x:x.split(','))
```

```python
#extracting all the sellers names and placing them in a new column for each
table['Seller name 1']=table['sellers'].apply(lambda x:str(x[0]).replace('Seller_name_1=>',''))
table['Seller name 2']=table['sellers'].apply(lambda x:str(x[2:3]).replace('Seller_name_2=>',''))
table['Seller name 3']=table['sellers'].apply(lambda x:str(x[4:5]).replace('Seller_name_3=>',''))
table['Seller name 4']=table['sellers'].apply(lambda x:str(x[6:7]).replace('Seller_name_4=>',''))
table['Seller name 5']=table['sellers'].apply(lambda x:str(x[8:9]).replace('Seller_name_5=>',''))
table['Seller name 6']=table['sellers'].apply(lambda x:str(x[10:11]).replace('Seller_name_6=>',''))
table['Seller name 7']=table['sellers'].apply(lambda x:str(x[12:13]).replace('Seller_name_7=>',''))
table['Seller name 8']=table['sellers'].apply(lambda x:str(x[14:15]).replace('Seller_name_8=>',''))
table['Seller name 9']=table['sellers'].apply(lambda x:str(x[16:17]).replace('Seller_name_9=>',''))
table['Seller name 10']=table['sellers'].apply(lambda x:str(x[18:19]).replace('Seller_name_10=>',''))
```

```python
#extracting all the corresponding sellers prices and placing them in a new column for each 
table["Seller price 1"]=table['sellers'].apply(lambda x:str(str(x[1:2]).split('£')[1:2]))
table["Seller price 2"]=table['sellers'].apply(lambda x:str(str(x[3:4]).split('£')[1:2]))
table["Seller price 3"]=table['sellers'].apply(lambda x:str(str(x[5:6]).split('£')[1:2]))
table["Seller price 4"]=table['sellers'].apply(lambda x:str(str(x[7:8]).split('£')[1:2]))
table["Seller price 5"]=table['sellers'].apply(lambda x:str(str(x[9:10]).split('£')[1:2]))
table["Seller price 6"]=table['sellers'].apply(lambda x:str(str(x[11:12]).split('£')[1:2]))
table["Seller price 7"]=table['sellers'].apply(lambda x:str(str(x[13:14]).split('£')[1:2]))
table["Seller price 8"]=table['sellers'].apply(lambda x:str(str(x[15:16]).split('£')[1:2]))
table["Seller price 9"]=table['sellers'].apply(lambda x:str(str(x[17:18]).split('£')[1:2]))
table["Seller price 10"]=table['sellers'].apply(lambda x:str(str(x[19:20]).split('£')[1:2]))
```

```python
#to show the progress
table.head(3)
```

```python
#dropping the 'sellers' and 'product_information' columns as all the data required from those columns are now extracted
table=table.drop(['sellers','product_information'],axis=1)
#stripping further unwanted symbols from the table
table = table.applymap(lambda x: str(x).lstrip('|[\'\"],\*').rstrip('\*[\'\",]'))
```

```python
def null(x):
    if x=='nan':
        return NA
    elif x=='':
        return NA
    else:
        return x
     
table=table.applymap(null)
```

```python
#to show the progress
table.head(3)
```

```python
#further cleaning the data of 'Price(£)' to check if any of the values are not present in the desired manner
pat=re.compile('\d*.*\d*\-\d*.*\d*')
l=table['Price(£)']
for index in l:
    x= pat.search(str(index))
    if x:
        print(x.group())
```

```python
#rows containing price in the form of a range were replaced with their mean
pat=re.compile('\d*.*\d*\-\d*.*\d*')
l=[]
for index in table['Price(£)']:
    index= pat.search(str(index))
    if index:
        index=index.group()
        index=index.split('-')
        index=(float(index[0])+float(index[1]))/2
        l.append(index)
    else:
        l.append(index)
table['Price(£)']=pd.Series(l)
```

```python
#this shows that those patterns no longer exist
pat=re.compile('\d*.*\d*\-\d*.*\d*')
for index in table['Price(£)']:
    x= pat.match(str(index))
    if x:
        print(x.group())
```

```python
#converting the 'Age Group','Average Review Rating' to float and 'Number Available In Stock' to int
table['Age Group']=table['Age Group'].map(lambda x:float(x))
table['Average Review Rating']=table['Average Review Rating'].map(lambda x:float(x))
table['Number Available In Stock']=table['Number Available In Stock'].map(lambda x:int(x))
table['Price(£)']=table['Price(£)'].map(lambda x: float(x))
```
