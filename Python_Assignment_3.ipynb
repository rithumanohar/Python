```python
# the numpy,pandas and string packages are imported
import numpy as np
import pandas as pd
from string import ascii_uppercase
```

```python
#the function below ensures that the randomly generated grades which are above 100 are subracted by 10 units and made below 100
def sub_over_100(array):
    for i in range(len(array)):
        if array[i] >= 100:
            array[i]-=10
    return array
```

```python
#defining the grading scheme breakpoint
Sce=[-2.0,-1.0,0.0,1.0]
```

```python
#the my_grader1 function takes two arguments(Grades of sections(argument),Grading Scheme(keyword argument)) and calculates the normalized grades,letter grade and returns a dataframe of section,raw grade,normalized grade and letter grade
from collections import Counter
def my_grader1(*GS,**S):
    GS_ALL=[]
    Raw_Grade=[]
    Normalized_Grade=[]
    Section=[]
    Letter_Grade=[]
    Grade_list=['A','B','C','D','F']#defining the list of grades
    #the below for loop gets the grading breakpoints and assigns to temp
    for key in S:
        temp=S[key]
    #the below for loop concatenates the grades of all the sections provided in the argument GS list in GS_ALL
    for j in GS:
        GS_ALL.extend(j)
    #the below function call sub_over_100 ensures all the grades passed are below 100
    GS_ALL=sub_over_100(GS_ALL)
    #the functions np.mean calculates the mean of all grades and np.std calculates the standard deviation of all grades
    m=np.mean(GS_ALL)
    sd=np.std(GS_ALL)
    #the below for loop makes a list Raw_Grade of all raw grades obtained in GS_ALL and also calculates the normalized grades
    for k in GS_ALL:
        Raw_Grade.append(k)
        nm_grade=(k-m)/sd
        Normalized_Grade.append(nm_grade)
    #the below function generates the Section for each of the input Sections in the GS list argument
    for j in range(len(GS)):#for each of the list
        for l in range(len(GS[j])):#the section for each of the elements of the list
            Section.append(ascii_uppercase[j])#ascii_uppercase is used to generate the sections
    # the below for loop calculates the letter grade for the normalized grades based on the input grading scheme
    for i in Normalized_Grade:
        if(i>temp[3]):
            Letter_Grade.append(Grade_list[0])
        elif(i<temp[3] and i>temp[2]):
            Letter_Grade.append(Grade_list[1])
        elif(i<temp[2] and i>temp[1]):
            Letter_Grade.append(Grade_list[2])
        elif(i<temp[1] and i>temp[0]):
            Letter_Grade.append(Grade_list[3])
        elif(i<temp[0]):
            Letter_Grade.append(Grade_list[4])
    #the 'Letters' Contains the type of grades in Letter_Grade
    Letters=Counter(Letter_Grade).keys() 
    # the 'Frequency conatins the count of the 'Letters' frequency
    Frequency=Counter(Letter_Grade).values() 
    print("the grades and their frequency are",list(zip(Letters,Frequency)))
    #the below dictionary creates the data for the data frame with Raw_Grade,Section,Normalized_Grade and Letter_Grade
    data_for_df={'Raw_Grade':Raw_Grade,'Section':Section,'Normalized_Grade':Normalized_Grade,'Letter_Grade':Letter_Grade}
    #the below df contains the dataframe created for output
    df=pd.DataFrame(data_for_df)
    #the below indexing indexed the dataframe from 1 onwards
    df.index=df.index+1
    #the below print statements prints the first 10(df.head(10)) and last 10(df.tail(10))rows
    print(df.head(10))
    print(df.tail(10))
    return df # returns the dataframe as output
```

```python
#function call with the 3 lists A,B,C sections and keyword argument S with grading scheme Sce passed
my_grader1(A,B,C,S=Sce)
```

```python
#Concatenates the grades into a single 1-D array "Result" and shows the result
temp=np.concatenate((A,B),axis=0)
Result=np.concatenate((temp,C),axis=0)
Result
#Calculates the mean (m) and standard deviation (sd) of the single array and print the results.
m=np.mean(Result)
print("The mean is",m)
sd=np.std(Result)
print("The standard deviation is",sd)
#Normalizes the single array "Result" by subtracting the mean m from each grade and dividing by the standard deviation sd
for i in range(len(Result)):
    Result[i]-=m
    Result[i]/=sd
Result
```

```python
#The below code implements the below mentioned grading scheme and prints the report
#Grading Scheme
#    grade<-2-->F
#-2<=grade<-1-->D
#-1<=grade<0-->C
# 0<=grade<1-->B
#   grade>=1-->A
A=[]
B=[]
C=[]
D=[]
F=[]

for i in range(len(Result)):
    if Result[i] < -2:
        F.append(Result[i])
    elif Result[i] >= -2 and Result[i]<-1:
        D.append(Result[i])
    elif Result[i] >= -1 and Result[i]<0:
        C.append(Result[i])
    elif Result[i] >= 0 and Result[i]<1:
        B.append(Result[i])
    elif Result[i] >= 1:
        A.append(Result[i])
print("\n\nThe Report of number of Students who received each grade")
print("The number of students who received A are",len(A))
print("The number of students who received B are",len(B))
print("The number of students who received C are",len(C))
print("The number of students who received D are",len(D))
print("The number of students who received F are",len(F))
```
